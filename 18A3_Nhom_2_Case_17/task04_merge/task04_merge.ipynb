```python
import pandas as pd

faculty = pd.read_csv("../data_clean/faculty_info_clean.csv")
survey = pd.read_csv("../data_clean/course_survey_clean.csv")
comments = pd.read_csv("../data_clean/student_comments_clean.csv")

survey_faculty = pd.merge(
    survey,
    faculty,
    on="faculty_id",
    how="left"
)
missing_faculty = survey_faculty[
    survey_faculty["department"].isna()
]

print(
    "Số faculty_id không có trong faculty_info:",
    missing_faculty["faculty_id"].nunique()
)
full_data = pd.merge(
    survey_faculty,
    comments,
    on="survey_id",
    how="left"
)
no_comment = full_data[
    full_data["comment_text"].isna()
]

print(
    "Số survey_id không có comment_text:",
    no_comment["survey_id"].nunique()
)
survey_no_comment_record = survey[
    ~survey["survey_id"].isin(comments["survey_id"])
]

print(
    "Số survey_id không có bản ghi trong student_comments:",
    survey_no_comment_record.shape[0]
)
full_data.info()
full_data.head()
full_data.to_csv(
    "../ket_qua_phan_tich/faculty_survey_comments_merged.csv",
    index=False
)
```
