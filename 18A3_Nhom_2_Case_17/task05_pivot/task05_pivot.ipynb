bảng pivot thể hiện ĐIỂM ĐÁNH GIÁ TRUNG BÌNH THEO KHOA × MỨC ĐỘ SINH VIÊN

```python
import pandas as pd
df = pd.read_csv("../data_clean/faculty_survey_comments_merged.csv")
df.head()
```

```python

pivot_rating = pd.pivot_table(
    df,
    values="rating_score",
    index="department",
    columns="student_level",
    aggfunc="mean"
)

pivot_rating = pivot_rating.round(2)
pivot_rating
```

COMMENT TÍCH CỰC

```python
positive_df = df[df["sentiment_label"] == "Tích cực"]

pivot_positive = pd.pivot_table(
    positive_df,
    values="survey_id",
    index="department",
    columns="student_level",
    aggfunc="count"
)

pivot_positive
```

STACK / UNSTACK

```python
stacked_rating = pivot_rating.stack()
stacked_rating

unstacked_rating = stacked_rating.unstack()
unstacked_rating
```

### Nhận xét
- Điểm đánh giá trung bình của sinh viên năm trên nhìn chung cao hơn sinh viên năm đầu.
- Một số khoa có điểm thấp và ít bình luận tích cực ở nhóm sinh viên mới.
- Pivot table giúp so sánh trực quan giữa các khoa và nhóm sinh viên.
